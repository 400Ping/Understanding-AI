```python
import numpy as np
import pandas as pd

# Load the IRIS dataset
iris_data = pd.read_csv('Iris.csv')

# Extract features and species
features = iris_data[['sl', 'sw', 'pl', 'pw']].values
species = iris_data['Type'].unique()

# Create a 3D array to store the reshaped data
reshaped_data = np.zeros((3, 4, 25))

# Reshape the data for each species
for i, specie in enumerate(species):
    specie_data = iris_data[iris_data['Type'] == specie][['sl', 'sw', 'pl', 'pw']].values[:25]
    reshaped_data[i] = specie_data.T

# Calculate averages, maximums, and minimums
for i, specie in enumerate(species):
    print(f"\nStatistics for {specie}:")
    
    # Averages
    averages = np.mean(reshaped_data[i], axis=1)
    print("Averages (sl, sw, pl, pw):", averages)
    
    # Maximums
    maximums = np.max(reshaped_data[i], axis=1)
    print("Maximums (sl, sw, pl, pw):", maximums)
    
    # Minimums
    minimums = np.min(reshaped_data[i], axis=1)
    print("Minimums (sl, sw, pl, pw):", minimums)

```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the IRIS dataset
iris_data = pd.read_csv('Iris.csv')

# Create the scatter plot
plt.figure(figsize=(10, 6))

# Plot each species with a different color
species = iris_data['Type'].unique()
colors = ['blue', 'green', 'red']

for i, specie in enumerate(species):
    specie_data = iris_data[iris_data['Type'] == specie]
    plt.scatter(specie_data['sl'], specie_data['pl'], 
                c=colors[i], label=specie, alpha=0.7)

# Customize the plot
plt.xlabel('x', fontsize=12)
plt.ylabel('y', fontsize=12)
plt.title('Iris Data', fontsize=14)
plt.legend()
plt.grid(True, linestyle='--', alpha=0.7)

# Show the plot
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the Pizza-simp.csv file
pizza_data = pd.read_csv('Pizza.csv')

# Create a figure with two subplots
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 6))

# Colors for each pizza brand
colors = ['red', 'blue', 'green', 'orange', 'purple']

# Scatter plot for Protein vs. Fat
for i, brand in enumerate(pizza_data['brand'].unique()):
    brand_data = pizza_data[pizza_data['brand'] == brand]
    ax1.scatter(brand_data['f'], brand_data['p'], c=colors[i], label=brand, alpha=0.7)

ax1.set_xlabel('Fat')
ax1.set_ylabel('Protein')
ax1.set_title('Protein vs. Fat in Pizza Brands')
ax1.legend()

# Scatter plot for Sugar vs. Sodium
for i, brand in enumerate(pizza_data['brand'].unique()):
    brand_data = pizza_data[pizza_data['brand'] == brand]
    ax2.scatter(brand_data['so'], brand_data['su'], c=colors[i], label=brand, alpha=0.7)

ax2.set_xlabel('Sodium')
ax2.set_ylabel('Sugar')
ax2.set_title('Sugar vs. Sodium in Pizza Brands')
ax2.legend()

# Adjust layout and display the plots
plt.tight_layout()
plt.show()
```
